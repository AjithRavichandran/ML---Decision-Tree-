```python
import pandas as pd
```

```python
dataset=pd.read_csv('50_Startups.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
independent
```

```python
dependent=dataset[['Profit']]
dependent
```

```python
from sklearn.model_selection import train_test_split 
```

```python
input_train,input_test,output_train,output_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor as DTR
regressor=DTR(criterion='friedman_mse', splitter='best',max_features='sqrt')
regressor=regressor.fit(input_train,output_train)
```

```python
from matplotlib import pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
predict_value=regressor.predict(input_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(output_test,predict_value)
r2
```

```python
import pickle
```

```python
Filename='Decision Tree - Friedman_mse.sav'
pickle.dump(regressor,open(Filename,'wb'))
```

```python
load=pickle.load(open('Decision Tree - Friedman_mse.sav','rb'))
```

```python
RD_Spend=float(input("Enter your R&D Spend:"))
Administration=float(input("Enter your Administration Spend:"))
Marketing_Spend=float(input("Enter your Market spent:"))
State_Florida=int(input("State Florida 0 or 1:"))
State_New_York=int(input("State New York 0 or 1:"))
```

```python
result=load.predict([[RD_Spend,Administration,Marketing_Spend,State_Florida,State_New_York]])
```
